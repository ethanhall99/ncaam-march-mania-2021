```python
import math
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Global Variables
minSeason = 2021

pd.options.mode.chained_assignment = None
```

```python
# read in files
MMasseyOrdinals = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/MMasseyOrdinals.csv')
MRegularSeasonDetailedResults = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/MRegularSeasonDetailedResults.csv')
MTeamConferences = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/MTeamConferences.csv')
Conferences = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/Conferences.csv')
MTeams = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/MTeams.csv')
MConferenceTourneyGames = pd.read_csv('/Users/ethanhall/Desktop/Data/ncaam-march-mania-2021/MDataFiles_Stage2/MConferenceTourneyGames.csv')
```

```python
# build teams dataset
dfTeamConf = MTeamConferences[MTeamConferences['Season'] == minSeason]
dfTeamConf = pd.merge(dfTeamConf, Conferences, on = 'ConfAbbrev', how = 'left')
dfTeamConf = pd.merge(dfTeamConf, MTeams, on = 'TeamID', how = 'left')
dfTeamConf = dfTeamConf[['TeamID'
                        ,'PowerFive']]


dfOppTeamConf = MTeamConferences[MTeamConferences['Season'] == minSeason]
dfOppTeamConf = pd.merge(dfOppTeamConf, Conferences, on = 'ConfAbbrev', how = 'left')
dfOppTeamConf = pd.merge(dfOppTeamConf, MTeams, on = 'TeamID', how = 'left')
dfOppTeamConf = dfOppTeamConf[['TeamID'
                                ,'PowerFive']]
dfOppTeamConf.columns = ['OppTeamID', 'OppPowerFive']


dfTeamConf.head(n=10)
```

```python
# Game details dataset
dfGame = MRegularSeasonDetailedResults[MRegularSeasonDetailedResults['Season'] == minSeason]
dayAdj = min(dfGame['DayNum']) - 1

gameCols = ["DayNum", "TeamID", "OppTeamID", "WinLoss", "PF", "PA", 
            "NumOT", "FGM", "FGA", "FGM3", "FGA3", "FTM", "FTA", 
            "OR", "DR", "Ast", "TO", "Stl", "Blk", "Fouls"]

dfGameWinner = dfGame[["DayNum", "WTeamID", "LTeamID", "WScore", "LScore", 
                       "NumOT", "WFGM", "WFGA", "WFGM3", "WFGA3", "WFTM", "WFTA", 
                       "WOR", "WDR", "WAst", "WTO", "WStl", "WBlk", "WPF"]]
dfGameWinner.insert(loc = 3, column = 'WinLoss', value = 1)
dfGameWinner.columns = gameCols

dfGameLoser = dfGame[["DayNum", "LTeamID", "WTeamID", "LScore", "WScore", 
                      "NumOT", "LFGM", "LFGA", "LFGM3", "LFGA3", "LFTM", "LFTA", 
                      "LOR", "LDR", "LAst", "LTO", "LStl", "LBlk", "LPF"]]
dfGameLoser.insert(loc = 3, column = 'WinLoss', value = 0)
dfGameLoser.columns = gameCols


dfGameDetails = dfGameWinner.append(dfGameLoser)
dfGameDetails = pd.merge(dfOppTeamConf, dfGameDetails, on = 'OppTeamID', how = 'left')
dfGameDetails = pd.merge(dfTeamConf, dfGameDetails, on = 'TeamID', how = 'left')

dfGameDetails['DayNum'] = dfGameDetails['DayNum'] - dayAdj
dfGameDetails.insert(loc = 8, column = 'WeekNum', value = np.ceil(dfGameDetails['DayNum']/7).astype(int))

print(len(dfGameDetails.index)) #7710
dfGameDetails.head(n=10)
```

```python
# Summarize Game Stats
dfSeasonStats = dfGameDetails.groupby(['TeamID', 'PowerFive']).agg(
    GP = pd.NamedAgg(column = "WinLoss", aggfunc = 'size'),
    Wins = pd.NamedAgg(column = "WinLoss", aggfunc = sum),
    Loss = pd.NamedAgg(column = "WinLoss", aggfunc = min),
    PF = pd.NamedAgg(column = "PF", aggfunc = 'mean'),
    PA = pd.NamedAgg(column = "PA", aggfunc = 'mean'),
    FGPct = pd.NamedAgg(column = "FGM", aggfunc = 'mean'),
    FGM = pd.NamedAgg(column = "FGM", aggfunc = sum),
    FGA = pd.NamedAgg(column = "FGA", aggfunc = sum),
    FG3Pct = pd.NamedAgg(column = "FGM3", aggfunc = 'mean'),
    FGM3 = pd.NamedAgg(column = "FGM3", aggfunc = sum),
    FGA3 = pd.NamedAgg(column = "FGA3", aggfunc = sum),
    FTPct = pd.NamedAgg(column = "FTM", aggfunc = 'mean'),
    FTM = pd.NamedAgg(column = "FTM", aggfunc = sum),
    FTA = pd.NamedAgg(column = "FTA", aggfunc = sum),
    Reb = pd.NamedAgg(column = "OR", aggfunc = sum),
    OReb = pd.NamedAgg(column = "OR", aggfunc = 'mean'),
    DReb = pd.NamedAgg(column = "DR", aggfunc = 'mean'),
    Ast = pd.NamedAgg(column = "Ast", aggfunc = 'mean'),
    TO = pd.NamedAgg(column = "TO", aggfunc = 'mean'),
    Stl = pd.NamedAgg(column = "Stl", aggfunc = 'mean'),
    Blk = pd.NamedAgg(column = "Blk", aggfunc = 'mean')
)

dfSeasonStats['Loss'] =  dfSeasonStats['GP'] - dfSeasonStats['Wins']
dfSeasonStats['FGPct'] = (dfSeasonStats['FGM']/dfSeasonStats['FGA'])*100
dfSeasonStats['FG3Pct'] = (dfSeasonStats['FGM3']/dfSeasonStats['FGA3'])*100
dfSeasonStats['FTPct'] = (dfSeasonStats['FTM']/dfSeasonStats['FTA'])*100
dfSeasonStats['Reb'] = dfSeasonStats['OReb'] + dfSeasonStats['DReb']


dfSeasonStats.reset_index(inplace = True)
dfSeasonStats.head(n=10)
```
